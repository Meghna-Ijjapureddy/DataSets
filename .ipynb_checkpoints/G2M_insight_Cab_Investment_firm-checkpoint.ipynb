```python
# Hypothesis 3: Is payment mode affecting the cab selection?

df_hype3 = Merge3.groupby(by=["Company","Payment_Mode"]).size().unstack()
print(df_hype3)

df_hype3.plot.bar()
#sns.boxplot(x="Company", y="Payment_Mode", data=df_hype3)
```

```python
# Calculate the percentage of card to cash for each company
df_hype3['Percentage_Card'] = df_hype3.apply(lambda x: round(x['Card'] / (x['Card'] + x['Cash']) * 100, 2), axis=1)
df_hype3['Percentage_Cash'] = df_hype3.apply(lambda x: round(x['Cash'] / (x['Card'] + x['Cash']) * 100, 2), axis=1)

# Print the results
print(df_hype3)
```

```python
# Get the percentages for each payment mode by cab type
pink_cab_perc = df_hype3.loc['Pink Cab', ['Percentage_Card', 'Percentage_Cash']]
yellow_cab_perc = df_hype3.loc['Yellow Cab', ['Percentage_Card', 'Percentage_Cash']]

# Plot the pie chart for Pink Cab
plt.pie(pink_cab_perc, labels=pink_cab_perc.index, autopct='%1.1f%%', colors=['tab:blue', 'tab:orange'])
plt.title('Pink Cab Payment Mode Percentage')
plt.show()

# Plot the pie chart for Yellow Cab
plt.pie(yellow_cab_perc, labels=yellow_cab_perc.index, autopct='%1.1f%%', colors=['tab:blue', 'tab:orange'])
plt.title('Yellow Cab Payment Mode Percentage')
plt.show()
```

## Hypothesis 3: Is payment mode affecting the cab selection?

####        Considering the graph above, there are no majors observed which proves that one payment is prefered or is much feasible mode of payment in a specific cab company. 
####        Hence, there is no product strategy defined in terms of Payment Mode in each cab company.

```python
from mpl_toolkits.mplot3d import Axes3D 
from sklearn.preprocessing import LabelEncoder

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 6))
fig.subplots_adjust(wspace=0.4)

# Define colors for each category
colors = {'2-4': 'pink', '5-10': 'orange', '10-100': 'blue', '>100': 'green'}

# Plot 1
retention_df = Merge3[Merge3['Company'] == 'Pink Cab'].groupby('Customer ID').size()
retention_df = pd.cut(retention_df, [0, 2, 5, 10, 100], labels=['2-4', '5-10', '10-100', '>100'])
counts = retention_df.value_counts()
pie_chart1 = counts.plot.pie(autopct='%1.1f%%', ax=ax1, colors=[colors[x] for x in counts.index])
ax1.set_title('Pink Cab Customer Retention')

# Plot 2
retention_df = Merge3[Merge3['Company'] == 'Yellow Cab'].groupby('Customer ID').size()
retention_df = pd.cut(retention_df, [0, 2, 5, 10, 100], labels=['2-4', '5-10', '10-100', '>100'])
counts = retention_df.value_counts()
pie_chart2 = counts.plot.pie(autopct='%1.1f%%', ax=ax2, colors=[colors[x] for x in counts.index])
ax2.set_title('Yellow Cab Customer Retention')

plt.show()
```

```python
# Hypothesis 9: YOY growth rate increases in new customer users.

hype9 = pd.DataFrame({"Year": Merge3["Year"],
                    "Customer ID": Merge3["Customer ID"],
                    "Company": Merge3["Company"]})
year3 = hype9.groupby(by=["Year"])['Customer ID'].unique()
# print(year3)

count1 = 0
count2 = 0
for i, year in enumerate(year3.index):
    if year == 2017:
        for customer_id in year3.loc[year]:
            if customer_id not in year3.iloc[i-1]:
                count1 += 1
    if year == 2018:
        for customer_id in year3.loc[year]:
            if customer_id not in year3.iloc[i-1]:
                count2 += 1
print(count1)
print(count2)
```

```python
# Hypothesis 12: Number of users switched form Pink to Yellow Cabs and vice versa

# Creating a new dataframe to count the number of users who switched from Pink to Yellow cab and vice versa
switch_df = pd.DataFrame(columns=['Year', 'Pink to Yellow', 'Yellow to Pink'])

# Looping through each year and counting the number of users who switched between the cabs
for year in sorted(Merge3['Year'].unique()):
    year_df = Merge3[Merge3['Year'] == year]
    pink_to_yellow = year_df[(year_df['Company'] == 'Pink Cab') & (year_df['Customer ID'].isin(year_df[year_df['Company'] == 'Yellow Cab']['Customer ID']))]['Customer ID'].nunique()
    yellow_to_pink = year_df[(year_df['Company'] == 'Yellow Cab') & (year_df['Customer ID'].isin(year_df[year_df['Company'] == 'Pink Cab']['Customer ID']))]['Customer ID'].nunique()
    switch_df = switch_df.append({'Year': year, 'Pink to Yellow': pink_to_yellow, 'Yellow to Pink': yellow_to_pink}, ignore_index=True)

# Printing the switch dataframe
print(switch_df)
```

```python
# Hypothesis 12: Number of users switched form Pink to Yellow Cabs and vice versa
####TAKE UNIQUE SET OF CID OF 2016 FOR PINK 1,2,3 AND YELLOW CABS (set(list)- gives unique elements output)
####TAKE UNIQUE SET OF CID OF 2017 FOR PINK AND YELLOW CABS 2,3,4
#### Subtract set of cid in 2016 and 2017 
#### pink 2016 -yellow 2017 will give 1
### and vice versa


# Creating a new dataframe to count the number of users who switched from Pink to Yellow cab and vice versa
switch_df = pd.DataFrame(columns=['Year', 'Pink to Yellow', 'Yellow to Pink'])

# Looping through each year and counting the number of users who switched between the cabs
for years in sorted(Merge3['Year'].unique()):
    year_df = Merge3[Merge3['Year'] == years]
    year1_df = Merge3[Merge3['Year'] == years+1]
    pink_to_yellow = year_df[(year_df['Company'] == 'Pink Cab') & ~(year_df['Customer ID'].isin(year1_df[year1_df['Company'] == 'Pink Cab']['Customer ID']))]['Customer ID'].nunique()
    yellow_to_pink = year_df[(year_df['Company'] == 'Yellow Cab') & ~(year_df['Customer ID'].isin(year1_df[year1_df['Company'] == 'Yellow Cab']['Customer ID']))]['Customer ID'].nunique()
    switch_df = switch_df.append({'Year': years, 'Pink to Yellow': pink_to_yellow, 'Yellow to Pink': yellow_to_pink}, ignore_index=True)

# Printing the switch dataframe
print(switch_df)
```

```python
# Hypothesis 12: Number of users switched form Pink to Yellow Cabs and vice versa
####TAKE UNIQUE SET OF CID OF 2016 FOR PINK 1,2,3 AND YELLOW CABS (set(list)- gives unique elements output)
####TAKE UNIQUE SET OF CID OF 2017 FOR PINK AND YELLOW CABS 2,3,4
#### Subtract set of cid in 2016 and 2017 
#### pink 2016 -yellow 2017 will give 1
### and vice versa


Pink = Merge3[Merge3["Company"]== "Pink Cab"]
Pink_CID_Unique = Pink.groupby("Year")["Customer ID"].unique()


 
Yellow = Merge3[Merge3["Company"]== "Yellow Cab"]
Yellow_CID_Unique = Yellow.groupby("Year")["Customer ID"].unique()

Pink_CID_Unique_2016 = set(Pink_CID_Unique[2016])
Pink_CID_Unique_2017 = set(Pink_CID_Unique[2017])
Pink_CID_Unique_2018 = set(Pink_CID_Unique[2018])

Yellow_CID_Unique_2016 = set(Yellow_CID_Unique[2016])
Yellow_CID_Unique_2017 = set(Yellow_CID_Unique[2017])
Yellow_CID_Unique_2018 = set(Yellow_CID_Unique[2018])

Pink_to_Yellow = Pink_CID_Unique_2016 - (Pink_CID_Unique_2017 | Pink_CID_Unique_2018)
print (len(Pink_CID_Unique_2016))
print (len(Pink_to_Yellow))
```

```python
# To check the churn rate
pink_customers_2016 = set(Pink_CID_Unique[2016])
pink_customers_2017 = set(Pink_CID_Unique[2017])
pink_customers_2018 = set(Pink_CID_Unique[2018])
pink_customers_2017_2018 = pink_customers_2017.union(pink_customers_2018)
pink_churned_customers = pink_customers_2016.difference(pink_customers_2017_2018)
yellow_customers_2018 = set(Yellow_CID_Unique[2018])
yellow_customers_2019 = set(Yellow_CID_Unique[2019])
churned_to_yellow = pink_churned_customers.intersection(yellow_customers_2018).intersection(yellow_customers_2019)
num_churned_to_yellow = len(churned_to_yellow)
print(num_churned_to_yellow)
```
