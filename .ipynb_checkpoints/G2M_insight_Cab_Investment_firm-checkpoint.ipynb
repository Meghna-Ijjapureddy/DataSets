```python
# Hypothesis 3: Is payment mode affecting the cab selection?

df_hype3 = Merge3.groupby(by=["Company","Payment_Mode"]).size().unstack()
print(df_hype3)

df_hype3.plot.bar()
#sns.boxplot(x="Company", y="Payment_Mode", data=df_hype3)
```

```python
# Calculate the percentage of card to cash for each company
df_hype3['Percentage_Card'] = df_hype3.apply(lambda x: round(x['Card'] / (x['Card'] + x['Cash']) * 100, 2), axis=1)
df_hype3['Percentage_Cash'] = df_hype3.apply(lambda x: round(x['Cash'] / (x['Card'] + x['Cash']) * 100, 2), axis=1)

# Print the results
print(df_hype3)
```

```python
# Get the percentages for each payment mode by cab type
pink_cab_perc = df_hype3.loc['Pink Cab', ['Percentage_Card', 'Percentage_Cash']]
yellow_cab_perc = df_hype3.loc['Yellow Cab', ['Percentage_Card', 'Percentage_Cash']]

# Plot the pie chart for Pink Cab
plt.pie(pink_cab_perc, labels=pink_cab_perc.index, autopct='%1.1f%%', colors=['tab:blue', 'tab:orange'])
plt.title('Pink Cab Payment Mode Percentage')
plt.show()

# Plot the pie chart for Yellow Cab
plt.pie(yellow_cab_perc, labels=yellow_cab_perc.index, autopct='%1.1f%%', colors=['tab:blue', 'tab:orange'])
plt.title('Yellow Cab Payment Mode Percentage')
plt.show()
```

## Hypothesis 3: Is payment mode affecting the cab selection?

####        Considering the graph above, there are no majors observed which proves that one payment is prefered or is much feasible mode of payment in a specific cab company. 
####        Hence, there is no product strategy defined in terms of Payment Mode in each cab company.

```python
# Hypothesis 6: Customer retension based number of times each customer using the cab
from mpl_toolkits.mplot3d import Axes3D 
from sklearn.preprocessing import LabelEncoder
Retension_df=pd.DataFrame({"Customer ID": Merge3["Customer ID"],
                          "Company": Merge3["Company"]})

customer_retension2 = Retension_df.groupby(by=["Company", "Customer ID"]).filter(lambda x: x['Customer ID'].count() >= 2)
customer_retension5 = Retension_df.groupby(by=["Customer ID","Company"]).filter(lambda x: x['Customer ID'].count() >= 5)
customer_retension10 = Retension_df.groupby(by=["Customer ID","Company"]).filter(lambda x: x['Customer ID'].count() >= 10)

#print (customer_retension2)

Retension2 = customer_retension2.groupby(by=["Customer ID","Company"]).size()
Retension5 = customer_retension5.groupby(by=["Customer ID","Company"]).size()
Retension10 = customer_retension10.groupby(by=["Customer ID","Company"]).size()


plt.figure()
sns.scatterplot(x=Retension2.index.get_level_values(0), y=Retension2, color='red')
sns.scatterplot(x=Retension5.index.get_level_values(0), y=Retension5, color='green')
sns.scatterplot(x=Retension10.index.get_level_values(0), y=Retension10, color='blue')
plt.xlabel('Customer ID')
plt.ylabel('Retention')
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D 
from sklearn.preprocessing import LabelEncoder

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 6))
fig.subplots_adjust(wspace=0.4)

# Define colors for each category
colors = {'2-4': 'pink', '5-10': 'orange', '10-100': 'blue', '>100': 'green'}

# Plot 1
retention_df = Merge3[Merge3['Company'] == 'Pink Cab'].groupby('Customer ID').size()
retention_df = pd.cut(retention_df, [0, 2, 5, 10, 100], labels=['2-4', '5-10', '10-100', '>100'])
counts = retention_df.value_counts()
pie_chart1 = counts.plot.pie(autopct='%1.1f%%', ax=ax1, colors=[colors[x] for x in counts.index])
ax1.set_title('Pink Cab Customer Retention')

# Plot 2
retention_df = Merge3[Merge3['Company'] == 'Yellow Cab'].groupby('Customer ID').size()
retention_df = pd.cut(retention_df, [0, 2, 5, 10, 100], labels=['2-4', '5-10', '10-100', '>100'])
counts = retention_df.value_counts()
pie_chart2 = counts.plot.pie(autopct='%1.1f%%', ax=ax2, colors=[colors[x] for x in counts.index])
ax2.set_title('Yellow Cab Customer Retention')

plt.show()
```

```python
# Hypothesis 9: YOY growth rate increases in new customer users.

hype9 = pd.DataFrame({"Year": Merge3["Year"],
                    "Customer ID": Merge3["Customer ID"],
                    "Company": Merge3["Company"]})
year3 = hype9.groupby(by=["Year"])['Customer ID'].unique()
# print(year3)

count1 = 0
count2 = 0
for i, year in enumerate(year3.index):
    if year == 2017:
        for customer_id in year3.loc[year]:
            if customer_id not in year3.iloc[i-1]:
                count1 += 1
    if year == 2018:
        for customer_id in year3.loc[year]:
            if customer_id not in year3.iloc[i-1]:
                count2 += 1
print(count1)
print(count2)
```

## Hypothesis 9: YOY growth rate increases in new customer users.

#### Though addition of new customers have reduced on a whole by a bit, but there is still a noteworty amount of new customers still getting added to the cab industry.

```python
# Hypothesis 12: Churn rate for Pink and Yellow Cabs

Pink = Merge3[Merge3["Company"]== "Pink Cab"]
Pink_CID_Unique = Pink.groupby("Year")["Customer ID"].unique()


Yellow = Merge3[Merge3["Company"]== "Yellow Cab"]
Yellow_CID_Unique = Yellow.groupby("Year")["Customer ID"].unique()

Pink_CID_Unique_2016 = set(Pink_CID_Unique[2016])
Pink_CID_Unique_2017 = set(Pink_CID_Unique[2017])
Pink_CID_Unique_2018 = set(Pink_CID_Unique[2018])

Yellow_CID_Unique_2016 = set(Yellow_CID_Unique[2016])
Yellow_CID_Unique_2017 = set(Yellow_CID_Unique[2017])
Yellow_CID_Unique_2018 = set(Yellow_CID_Unique[2018])

Leave_Pink = Pink_CID_Unique_2016 - (Pink_CID_Unique_2017 | Pink_CID_Unique_2018)
Pink_Churn_rate = len(Leave_Pink)/len(Pink_CID_Unique_2016)*100
print (Pink_Churn_rate)

Leave_Yellow = Yellow_CID_Unique_2016 - (Yellow_CID_Unique_2017 | Yellow_CID_Unique_2018)
Yellow_Churn_rate = len(Leave_Yellow)/len(Yellow_CID_Unique_2016)*100
print (Yellow_Churn_rate)
```

```python
# Plot the bar graph
fig, ax = plt.subplots()
ax.bar(["Pink Cab", "Yellow Cab"], [Pink_Churn_rate, Yellow_Churn_rate], color=["pink", "Gold"])
ax.set_ylabel("Churn Rate (%)")
ax.set_title("Churn Rate Comparison - Pink vs Yellow Cabs for the year 2016-17")

plt.show()
```

```python
# Churn rate form 2017 to 2018
Leave_Pink_2018 = Pink_CID_Unique_2017 - (Pink_CID_Unique_2018)
Pink_Churn_rate2018 = len(Leave_Pink_2018)/len(Pink_CID_Unique_2017)*100
print (Pink_Churn_rate2018)

Leave_Yellow_2018 = Yellow_CID_Unique_2017 - (Yellow_CID_Unique_2018)
Yellow_Churn_rate2018 = len(Leave_Yellow_2018)/len(Yellow_CID_Unique_2017)*100
print (Yellow_Churn_rate2018)
```

```python
# Plot the bar graph
fig, ax = plt.subplots()
ax.bar(["Pink Cab", "Yellow Cab"], [Pink_Churn_rate2018, Yellow_Churn_rate2018], color=["pink", "Gold"])
ax.set_ylabel("Churn Rate (%)")
ax.set_title("Churn Rate Comparison - Pink vs Yellow Cabs for the year 2017-18")

plt.show()
```

## Hypothesis 12: Churn rate for Pink and Yellow Cabs

#### Overall, the Churn rate for Pink Cab is higher than Churn rate for Yellow Cabs, as per above calculations.
